## Análises dos dados de Localidade dos Alunos da UFES

Este notebook apresenta algumas análises realizadas a partir das tabelas da camada Gold, construídas durante o pipeline de Engenharia de Dados.

```python
from pyspark.sql import functions as F

df_municipios = spark.table("workspace.gold.alunos_por_municipio")
df_cursos = spark.table("workspace.gold.alunos_por_curso")
df_estados = spark.table("workspace.gold.alunos_por_estado")
df_anos = spark.table("workspace.gold.alunos_por_ano")
df_top_curso = spark.table("workspace.gold.top_municipio_por_curso")
df_alunos_2025_curso = spark.table("workspace.gold.alunos_2025_curso")
```

```python
display(df_municipios.limit(10))
```

- Os dados mostram que São Mateus concentra a maior quantidade de estudantes, seguido por Conceição da Barra e Linhares. Isso é esperado devido à proximidade geográfica com o campus da UFES em São Mateus.

```python
display(df_cursos)
```

- Atualmente o curso com mais alunos é o de Ciência da Computação, que pode significar um aumento atual ou uma permanência dos estudantes para concluir o curso.

```python
display(df_alunos_2025_curso)
```

- Mas ao conferir a quantidade de alunos ingressantes em 2025, vemos que o curso de Ciência da Computação, Agronomia e Engenharia da Computação tem mais ingressantes, mostrando que a quantidade alta de alunos nesse curso é pela demanda atual.

```python
display(df_estados)
```

- A maioria dos estudantes é proveniente do Espírito Santo, mas há participação alta de estados vizinhos como Bahia e Minas Gerais e alunos de váriso outros anos.

```python
display(df_anos)
```

- Aqui podemos ver o aumento de alunos em anos mais recentes, o que faz sentido, pois os demais ou se formaram ou saíram do curso (os dados da tabela não permitem saber essa informação).

```python
display(df_top_curso)
```

- Para a maioria dos cursos, o endereço dos estudantes foi São Mateus
