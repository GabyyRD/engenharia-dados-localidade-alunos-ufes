### Objetivo

Realizar a ingestão dos dados brutos de alunos da UFES

Arquivo:
Localidade_Alunos.csv

Destino:
workspace.bronze.alunos_raw

```python
spark
```

```python
print(type(spark))
spark.version
```

```python
from pyspark.sql import functions as F
```

```python
df_alunos = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ",") \
    .load("/Volumes/workspace/default/dados/Localidade_Alunos.csv")

df_alunos.printSchema() #colar resultado aqui depois
display(df_alunos.limit(10))
df_alunos.count() #1721
```

```python
df_alunos.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.alunos_raw")
```

```python
#Verificando se foi feita a escrita no delta
spark.sql("SHOW TABLES IN bronze").show()
df_bronze = spark.table("bronze.alunos_raw")

display(df_bronze)

spark.sql("SELECT current_catalog(), current_schema()").show()
```
